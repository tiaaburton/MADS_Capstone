```python
ticker = "AAPL"

model_df = ticker_regression.retrieve_single_ticker_model_results_from_mongo(ticker)

model_df.drop(labels=["_id"], axis=1, inplace=True)
# model_df.rename(columns={"index": "Date"}, inplace=True)

model_df["Date"] = pd.to_datetime(model_df["Date"])
model_df.drop(["Close"], axis=1, inplace=True)
model_df.set_index("Date", inplace=True)

# print(model_df)

yahoo_df = yahoo.retrieve_company_stock_price_from_mongo(ticker)
yahoo_df.drop(labels=["_id"], axis=1, inplace=True)
yahoo_df["Date"] = pd.to_datetime(yahoo_df["Date"])
yahoo_df.set_index("Date", inplace=True)
# print(yahoo_df)
# # model_df.sort_values(["Date"], ascending=True, inplace=True)
# combined_df = yahoo_df.concat(model_df, left_index=True, right_index=True)
combined_df = yahoo_df.merge(model_df, how="right", left_index=True, right_index=True)
combined_df.reset_index(inplace=True)
combined_df.sort_values(["Date"], ascending=True, inplace=True)
px.line(combined_df, x='Date', y=['Close','prediction'])
```
