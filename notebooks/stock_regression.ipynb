```python
import plotly.graph_objs as go

ticker = "GE"

model_df = ticker_regression.retrieve_single_ticker_model_results_from_mongo(ticker)

fig = go.Figure([
    go.Scatter(
        name='Close Price',
        x=model_df['Date'],
        y=model_df['Close'],
        mode='lines',
        line=dict(color='rgb(31, 119, 180)'),
    ),
    go.Scatter(
        name='Prediction',
        x=model_df['Date'],
        y=model_df['prediction_upper'],
        mode='lines',
        marker=dict(color="#444"),
        line=dict(width=0),
        showlegend=False
    ),
    go.Scatter(
        name='Prediction',
        x=model_df['Date'],
        y=model_df['prediction_lower'],
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines',
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty',
        showlegend=True
    )
])
fig.update_layout(
    yaxis_title='Stock Price',
    hovermode="x"
)
fig.show()
```
